# Дані з таблиці таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
data = table[0]
print(data)
```

# Завдання №5
Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
numeric_columns = data.columns[1:]
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
filled_data = data.fillna({
    '1950': data['1950'].mean(),
    '1960': data['1960'].mean(),
    '1970': data['1970'].mean(),
    '1990': data['1990'].mean(),
    '2000': data['2000'].mean(),
    '2012': data['2012'].mean(),
    '2014': data['2014'].mean(),
    '2019': data['2019'].mean()
}).round(1)
print(filled_data)
```

# Завдання №9
Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні


```python
mean_data = data['2019'].mean()
mask = data['2019'] > mean_data

regions = data.loc[mask, 'Регіон'].tolist()
regions
```

# Завдання №10
У якому регіоні була найвища народжуваність у 2014 році?

```python
max_value = filled_data["2014"].idxmax()
result  = data.loc[max_value, "Регіон"]
print(f"Лідером по народжуванності в 2014 році була {result} область")
```

# Завдання №11
Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
import matplotlib.pyplot as plt


data_cleaned = data.dropna()
regions = data_cleaned["Регіон"]
birth_rates_2019 = data_cleaned["2019"]


plt.figure(figsize=(10, 6))
plt.bar(regions, birth_rates_2019, color='m')


plt.title('Рівень народжуваності за регіонами в 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
