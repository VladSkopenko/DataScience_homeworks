# Дані з таблиці таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
import pandas as pd
import numpy as np

url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"

table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', decimal=',', thousands='.')
data = table[0]
print(data)
```

# Завдання №5
Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
numeric_columns = data.columns[1:]
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
filled_data = data.fillna({
    '1950': data['1950'].mean(),
    '1960': data['1960'].mean(),
    '1970': data['1970'].mean(),
    '1990': data['1990'].mean(),
    '2000': data['2000'].mean(),
    '2012': data['2012'].mean(),
    '2014': data['2014'].mean(),
    '2019': data['2019'].mean()
}).round(1)
print(filled_data)
```

# Завдання №11
Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році


```python
import matplotlib.pyplot as plt


data_cleaned = data.dropna()
regions = data_cleaned["Регіон"]
birth_rates_2019 = data_cleaned["2019"]


plt.figure(figsize=(10, 6))
plt.bar(regions, birth_rates_2019, color='m')


plt.title('Рівень народжуваності за регіонами в 2019 році')
plt.xlabel('Регіон')
plt.ylabel('Рівень народжуваності')

plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

# Завдання №12
додайте від 3 до 5 графіків функцій різного типу на ваш вибір

```python
import matplotlib.pyplot as plt


regions = filled_data['Регіон']
birth_rates = filled_data.drop('Регіон', axis=1)
mean_birth_rate = birth_rates.mean(axis=1)


difference = birth_rates.sub(mean_birth_rate, axis=0)

plt.figure(figsize=(12, 8))
plt.barh(regions, difference.mean(axis=1), color='Indigo')
plt.axvline(0, color='gray', linewidth=0.8)
plt.title('Відмінності рівня рожденості від середнього за регіонами')
plt.xlabel('Відмінність від середнього')
plt.ylabel('Регіон')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd


odesa_birth_rate = filled_data[filled_data['Регіон'] == 'Одеська'].iloc[:, 1:].squeeze().astype(float)
kiev_birth_rate = filled_data[filled_data['Регіон'] == 'Київ'].iloc[:, 1:].squeeze().astype(float)


odesa_birth_rate.index = kiev_birth_rate.index = filled_data.columns[1:]


odesa_vs_kiev_birth_rate = pd.DataFrame({'Одеса': odesa_birth_rate, 'Київ': kiev_birth_rate})
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(odesa_vs_kiev_birth_rate.index, odesa_vs_kiev_birth_rate['Одеса'], marker='o', label='Одеса')
plt.plot(odesa_vs_kiev_birth_rate.index, odesa_vs_kiev_birth_rate['Київ'], marker='o', label='Київ')


plt.title('Зміна коефіцієнта народжуваності в Одесі та Києві з 1950 по 2019 рік')
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.legend()


plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data = filled_data

data_sorted = data.sort_values(by='2019', ascending=False)
plt.figure(figsize=(10, 6))

bars = plt.bar(data_sorted['Регіон'], data_sorted['2000'], color=['skyblue', 'lightgreen', 'salmon', 'orange', 'lightcoral'], edgecolor='black', alpha=0.7)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 0.05, round(height, 1), ha='center', va='bottom', fontsize=10)


plt.title('Лідери народжуванності в 2000 году', fontsize=16)
plt.xlabel('Регіон', fontsize=14)
plt.ylabel('Кількість народжень', fontsize=14)


plt.box(False)


plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.axhline(y=np.mean(data_sorted['2000']), color='gray', linestyle='--', linewidth=1)
plt.text(len(data_sorted['Регіон'])-0.5, np.mean(data_sorted['2000'])+0.1, 'Среднее: {:.1f}'.format(np.mean(data_sorted['2000'])), ha='right', va='bottom', fontsize=12, color='gray')
plt.xticks(rotation=90, fontsize=12)


plt.show()


```
