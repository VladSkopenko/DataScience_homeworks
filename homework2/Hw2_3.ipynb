Відповідь: Скільки книг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках (використовуйте | або функцію isin)? Відповідь:

```python
us_rate = data["user_rating"] == 4.9

date_data = data["year"].isin([2010, 2011])

result = (us_rate & date_data).sum()
print(f'Кількість ниг з рейтингом 4.9 потрапило до рейтингу у 2010 та 2011 роках: {result}')
```

## Тепер створіть новий датафрейм, який вміщатиме кількість книг для кожного з авторів (використовуйте функції groupby та agg, для підрахунку кількості використовуйте count). Не беріть усі стовпці, виберете тільки потрібні

```python
author_book_count = data.groupby("author").size().reset_index(name="Кількість книг")
print(f"Розмір фрейма {author_book_count.size}")
author_most_books = author_book_count["author"].iloc[author_book_count["Кількість книг"].idxmax()]
print(f"Автор з найбільщою кількістю книг {author_most_books}")
max_books_author_index = author_book_count["Кількість книг"].idxmax()
max_books_count = author_book_count.loc[max_books_author_index, "Кількість книг"]
print(f"Кількість книг у цього автора {max_books_count}")
```

## Тепер створіть другий датафрейм, який буде вміщати середній рейтинг для кожного автора (використовуйте функції groupby та agg, для підрахунку середнього значення використовуйте mean). Не беріть усі стовпці, виберете тільки потрібні

```python
authors_ratings = data.groupby("author")["user_rating"].mean().reset_index(name="midrate")
min_avg_rating_index = authors_ratings["midrate"].idxmin()
author = new_data.loc[min_avg_rating_index, "author"]
print(f"Автор з найменшим рейтингом {author}")
result = authors_ratings[authors_ratings["author"] == author]
result = result["midrate"]
print(f'Середній рейтинг цього автора {result}')
```

```python
import matplotlib.pyplot as plt


genre_prices = data.groupby('genre')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(genre_prices['genre'], genre_prices['price'], color='r')
plt.title('Залежність ціни від жанра')
plt.xlabel('Жанр')
plt.ylabel('Середня ціна')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
data_2009_2019 = data[(data['year'] >= 2009) & (data['year'] <= 2019)]
yearly_prices = data_2009_2019.groupby('year')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(yearly_prices['year'], yearly_prices['price'], marker='o', color='g', linestyle='-')
plt.title('Зміна цін на книги з 2009 по 2019 рік')
plt.xlabel('Рік')
plt.ylabel('Середня ціна')
plt.grid(True)
plt.xticks(yearly_prices['year'])
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.pie(data['genre'].value_counts(), labels=data['genre'].unique(), colors=['skyblue', 'lightgreen'], autopct='%1.1f%%', startangle=90)
plt.title('Розподіл жанрів книг')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns


author_book_count = data.groupby('author')['name'].count().reset_index()


top_authors = author_book_count.sort_values(by='name', ascending=False).head(5)


plt.figure(figsize=(10, 6))
sns.swarmplot(x='name', y='author', data=top_authors, size=10, palette='Set2')
plt.title('Топ-5 найпопулярніших авторів')
plt.xlabel('Кількість книг')
plt.ylabel('Автор')
plt.show()

```
